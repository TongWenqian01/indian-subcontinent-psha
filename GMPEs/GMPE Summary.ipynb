```python
import pandas as pd
from IPython.display import display
```

```python
gmpe_comparison_csv = '../Data/nath2011peak/Tables4-5.csv'
df_gmpes = pd.read_csv(gmpe_comparison_csv)
```

```python
for region in set(df_gmpes['Region']):
    indices = df_gmpes['Region'] == region
    weights = 2**(-df_gmpes[indices]['LLH'])
    df_gmpes.loc[indices, 'weight'] = (weights/weights.sum()).round(2)
    uniform_weight = 1./sum(indices)
    df_gmpes.loc[indices, 'DSI'] = (100*(weights - uniform_weight)/uniform_weight).round(1)
```

```python
for region in set(df_gmpes['Region']):
    indices = df_gmpes['Region'] == region
    print region
    display(df_gmpes[indices].drop('Region', axis=1))
    file_name = region + '.tex'
    df_gmpes[indices].drop('Region', axis=1).to_latex(file_name, index=False)
```
